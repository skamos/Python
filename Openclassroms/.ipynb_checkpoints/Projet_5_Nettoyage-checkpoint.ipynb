## Fonctions

```python
def post_to_words( post_text, stopwords_lem=True, html=True):
    # Function to convert a raw review to a string of words
    # The input is a single string (a raw movie review), and
    # the output is a single string (a preprocessed movie review)
    if html:
        post_text = BeautifulSoup(post_text).get_text()
        # 1. Remove HTML
    # 2. Remove non-letters
    #post_text = re.sub("[^a-zA-Z]", " ", post_text)
    post_text = re.sub("(\\n)|(\.\\n)|[;/\\\{}()*`':=\"|!?]", " ", post_text)
    post_text = re.sub("(\.\s+)|(\.$)|(,\s+)|(\s+#\s+)|(\s+<\s+)|(\s+>\s+)|(\s+\-\s+)", " ", post_text)
    post_text = re.sub("\s+", " ", post_text)
    #
    # 3. Convert to lower case, split into individual words
    words = post_text.lower().split()
    #
    if stopwords_lem :
        # 4. In Python, searching a set is much faster than searching
        # a list, so convert the stop words to a set
        stops = set(stopwords.words("english"))
        # 5. Remove stop words
        words = [w for w in words if not w in stops]
        # 6. Join the words back into one string separated by space,
        # and return the result.
        lemmatizer = WordNetLemmatizer()
        words = [lemmatizer.lemmatize(w) for w in words]
    return " ".join( words )
```

```python
def tags_for_question(question_id):
    return df_tags[df_tags['PostId'] == question_id].TagName.values

def add_tags_column(row):
    row['Tags_list'] = tags_for_question(row['Id_Post'])
    return row
```

```python
# Download the punkt tokenizer for sentence splitting
import nltk.data

# Load the punkt tokenizer
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

# Define a function to split a review into parsed sentences
def review_to_sentences( review, tokenizer, stopwords_lem=False ):
    # Function to split a review into parsed sentences. Returns a
    # list of sentences, where each sentence is a list of words
    #
    # 1. Use the NLTK tokenizer to split the paragraph into sentences
    raw_sentences = tokenizer.tokenize(review.strip())
    #
    # 2. Loop over each sentence
    sentences = []
    for raw_sentence in raw_sentences:
        # If a sentence is empty, skip it
        if len(raw_sentence) > 0:
            # Otherwise, call review_to_wordlist to get a list of words
            sentences.append( post_to_words( raw_sentence, stopwords_lem ))
    #
    # Return the list of sentences (each sentence is a list of words,
    # so this returns a list of lists
    return sentences
```
